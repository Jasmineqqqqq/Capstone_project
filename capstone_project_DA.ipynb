As we do not have standardised measurment values for serving size column, we have converted oz into grams and fl oz/cups into mili liotres. Some of the varied values include:  
16.2 oz (460 g)       
16.9 fl oz
1 carton (236 ml)           


```python
import re
def convert_to_numeric(serving_size):
    # First, try to extract and return grams if available
    grams_match = re.search(r'(\d+) g', serving_size)
    if grams_match:
        return float(grams_match.group(1))

    # Next, try to convert fluid ounces to milliliters
    fl_oz_match = re.search(r'(\d+) fl oz', serving_size)
    if fl_oz_match:
        fl_oz = float(fl_oz_match.group(1))
        ml = fl_oz * 29.5735  # Convert fl oz to ml
        return ml

    # Lastly, handle milliliters directly
    ml_match = re.search(r'(\d+) ml', serving_size)
    if ml_match:
        return float(ml_match.group(1))

    # Return None if no numeric conversion is possible
    return None

# Apply the conversion function to the 'Serving Size' column
df['Numeric Serving Size'] = df['Serving Size'].apply(convert_to_numeric)

# Convert the new column to float just to ensure consistency
df['Numeric Serving Size'] = df['Numeric Serving Size'].astype(float)

# Display the DataFrame to verify the changes
df[['Serving Size', 'Numeric Serving Size']].head(260)

```

```python
# Define the lower and upper percentiles to cap
lower_percentile = 0.01  # 1st percentile
upper_percentile = 0.99  # 99th percentile

# Compute the percentile values
lower_cap = df['Calories'].quantile(lower_percentile)
upper_cap = df['Calories'].quantile(upper_percentile)

# Apply capping
df['Calories_Capped'] = df['Calories'].clip(lower=lower_cap, upper=upper_cap)

# Display the first few rows to verify the changes
print(df[['Calories', 'Calories_Capped']].head())
```

```python

# Function to identify the highest and lowest calorie items
def find_extreme_calorie_items(df):
    highest_calorie_item = df.loc[df['Calories'].idxmax()]
    lowest_calorie_item = df.loc[df['Calories'].idxmin()]
    return highest_calorie_item, lowest_calorie_item

highest_calorie_item, lowest_calorie_item = find_extreme_calorie_items(df)

print("Highest Calorie Item:")
print(highest_calorie_item)
print("\nLowest Calorie Item:")
print(lowest_calorie_item)

```

```python
#trying it with median instead of mean as there are outliers
def calculate_average_nutritional_content(df):
    nutritional_columns = ['Calories', 'Dietary Fiber', 'Vitamin C %', 'Protein',
                           'Vitamin A %', 'Calcium %', 'Iron %']
    return df.groupby('Category')[nutritional_columns].median().sort_values(by='Calories', ascending=True)

average_nutritional_content = calculate_average_nutritional_content(df)

print("\nAverage Nutritional Content by Category (Sorted by Calories in Ascending Order using Median):")
print(average_nutritional_content)
```
